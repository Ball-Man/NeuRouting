```python
from utils.io import read_vrp

inst = read_vrp("../res/A-n32-k5.vrp", grid_dim=100)
inst.plot()
```

```python
from baselines import SCIPSolver

scipsolver = SCIPSolver()
scipsolver.solve(instance=inst, time_limit=10)
scipsolver.render()
```

```python
from baselines import OrToolsSolver

orsolver = OrToolsSolver()
orsolver.solve(instance=inst, time_limit=10)
orsolver.render()
```

```python
from baselines import LKHSolver

lkhsolver = LKHSolver(executable="../executables/LKH")
lkhsolver.solve(instance=inst)
lkhsolver.render()
```
