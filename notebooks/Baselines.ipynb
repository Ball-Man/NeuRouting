```python
from utils.io import read_vrp

inst = read_vrp("../res/A-n32-k5.vrp", grid_dim=100)
inst.plot()
```

```python
from baselines import SCIPSolver

scipsolver = SCIPSolver(instance=inst)
sol = scipsolver.solve(time_limit=10)
inst.plot(solution=sol)
```

```python
from baselines import OrToolsSolver

orsolver = OrToolsSolver(instance=inst)
sol = orsolver.solve(time_limit=10)
inst.plot(solution=sol)
```

```python
from baselines import LKHSolver

lkhsolver = LKHSolver(instance=inst, executable="../executables/LKH")
sol = lkhsolver.solve()
inst.plot(solution=sol)
```
